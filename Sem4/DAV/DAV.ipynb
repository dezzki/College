```python
#3
df = pd.DataFrame(np.random.randn(50, 3), columns=['A','B','C'])

for _ in range(int(0.1 * df.size)):
    i = np.random.randint(0, 50)
    j = np.random.randint(0, 3)
    df.iat[i, j] = np.nan
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna(axis=1, thresh=45)
```

```python
row = df.sum(axis=1).idxmax()
df = df.drop(row)
```

```python
df = df.drop_duplicates(subset=df.columns[0])
```

```python
print(df[df.columns[0]].corr(df[df.columns[1]]))
print(df[df.columns[1]].cov(df[df.columns[2]]))
```

```python
df['B_bins'] = pd.cut(df[df.columns[1]], bins=5)
print(df.head())
```
